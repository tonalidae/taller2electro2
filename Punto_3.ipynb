```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def matrix_trans(zi, zt):
    matrix = [[0.5*(1+zi/zt), 0.5*(1-zi/zt)],\
             [0.5*(1-zi/zt), 0.5*(1+zi/zt)]]
    return matrix


def matrix_propa(kd):
    matrix = [[np.exp(complex(0, -kd)), 0],\
             [0, np.exp(complex(0, kd))]]
    return matrix
    

T_12 = matrix_trans(1, 1/np.sqrt(2))
T_23 = matrix_trans(1, np.sqrt(2))

kd = np.linspace(0, np.pi, 100)

T = []
R = []
inpi2 = []

for ii in range (len(kd)):
    T_2 = matrix_propa(kd[ii])
    aux = np.matmul(T_12, T_2)
    matrix = np.matmul(aux, T_23)
    t = 1/(matrix[0][0])
    r = matrix[1][0]*t
    T.append(np.abs(t*t))
    R.append(np.abs(r*r))
    if (ii==int(len(kd)/2)):
        inpi2.append(np.abs(t*t))
        inpi2.append(np.abs(r*r))
    

plt.plot(kd, T, label='Transmitancia')
plt.plot(kd, R, label='Reflectancia')
plt.vlines(np.pi/2, 0, 1, colors='black', linestyles='dashed')
plt.plot([np.pi/2, np.pi/2], inpi2, "ro", )
plt.text(1,0.5,"$kd = \pi /2$", fontsize=12)
plt.text(1.6, 0.83, f"T={inpi2[0]:.2f}")
plt.text(1.6, 0.15, f"R={inpi2[1]:.2f}")
# plt.text(3.5,0.3, "Si el espesor de la l√°mina" 
#          "\n"
#          "es una cuarta parte de la"
#          "\n"
#          "longitud de onda, no hay"
#          "\n"
#          "variables libres en el problema."
#          "\n"
#          r"$kd=\pi/2$ corresponde a este caso.", fontsize=13)
plt.xlabel(r'$kd$')
plt.ylabel('Transmitancia y Reflectancia')
plt.grid()
plt.legend()
plt.savefig("Graficas2.png",bbox_inches="tight")
plt.show()
```

```python
def matrix_trans(zi, zt):
    """
    Calculate a 2x2 matrix based on the z_i z_t values.

    This function computes a 2x2 matrix using the input_val and target_val parameters.
    The matrix is calculated using the following formula:
    matrix[i][j] = 0.5 * (1 + (input_val / target_val) if i == j else 1 - (input_val / target_val))

    Parameters:
    input_val (float): The input value used in the matrix calculation.
    target_val (float): The target value used in the matrix calculation.

    Returns:
    list: A 2x2 matrix calculated based on the input_val and target_val parameters.

    Example:
    >>> matrix_trans(2, 4)
    [[0.75, 0.25], [0.25, 0.75]]
    """
    matrix = [[0.5 * (1 + zi/zt if i == j else 1 - zi/zt)
               for j in range(2)] for i in range(2)]
    return matrix
def matrix_propagation(kd):
    """
    This function generates a 2x2 matrix with diagonal elements representing
    the complex exponential of the input kd.

    Args:
    kd (float): The input value for calculating the complex exponential.

    Returns:
    list: A 2x2 matrix with diagonal elements as complex exponentials of the input kd.
    """
    
    # Calculate the complex exponential of -kd and kd
    exp_neg_kd = np.exp(complex(0, -kd))
    exp_kd = np.exp(complex(0, kd))

    # Create a 2x2 matrix with diagonal elements as complex exponentials
    matrix = [[exp_neg_kd, 0],
              [0, exp_kd]]

    return matrix
```

```python

T_12 = matrix_trans(1, 1/np.sqrt(2))
T_23 = matrix_trans(1, np.sqrt(2))

kd = np.linspace(0, np.pi, 100)

T = []
R = []
inpi2 = []

for ii in range (len(kd)):
    T_2 = matrix_propa(kd[ii])
    aux = np.matmul(T_12, T_2)
    matrix = np.matmul(aux, T_23)
    t = 1/(matrix[0][0])
    r = matrix[1][0]*t
    T.append(np.abs(t*t))
    R.append(np.abs(r*r))
    if (ii==int(len(kd)/2)):
        inpi2.append(np.abs(t*t))
        inpi2.append(np.abs(r*r))
    

plt.plot(kd, T, label='Transmitancia')
plt.plot(kd, R, label='Reflectancia')
plt.vlines(np.pi/2, 0, 1, colors='black', linestyles='dashed')
plt.plot([np.pi/2, np.pi/2], inpi2, "ro", )
plt.text(1,0.5,"$kd = \pi /2$", fontsize=12)
plt.text(1.6, 0.83, f"T={inpi2[0]:.2f}")
plt.text(1.6, 0.15, f"R={inpi2[1]:.2f}")
plt.xlabel(r'$kd$')
plt.ylabel('Transmitancia y Reflectancia')
plt.grid()
plt.legend()
plt.savefig("Graficas2.png",bbox_inches="tight")
plt.show()
```
